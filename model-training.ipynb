## Importing Libraries

```python
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import load_img
```

## Generate Train, Val, Test Dataset

```python
# Defining train-val-test generator
def train_val_test_generator(TRAIN_DIR, VAL_DIR, TEST_DIR):
    train_datagen = ImageDataGenerator(rescale=1/255.0,
                                       rotation_range=0.2,
                                       width_shift_range=0.2,
                                       height_shift_range=0.2,
                                       shear_range=0.2,
                                       zoom_range=0.2,
                                       horizontal_flip=True,
                                       fill_mode='nearest')

    train_generator = train_datagen.flow_from_directory(directory=VAL_DIR,
                                        target_size=(300, 300),
                                        class_mode='categorical',
                                        batch_size=10)

    val_datagen = ImageDataGenerator(rescale=1/255.0)

    val_generator = val_datagen.flow_from_directory(directory=VAL_DIR,
                                                    target_size=(300, 300),
                                                    class_mode='categorical',
                                                    batch_size=10)

    test_datagen = ImageDataGenerator(rescale=1/255.0)

    test_generator = test_datagen.flow_from_directory(directory=VAL_DIR,
                                                    target_size=(300, 300),
                                                    class_mode='categorical',
                                                    batch_size=10)

    return train_generator, val_generator, test_generator
```

### Define Dataset Directories

```python
dataset_dir = os.path.join(os.getcwd(), "Household_Waste")
train_dir = os.path.join(dataset_dir, "train")
val_dir = os.path.join(dataset_dir, "val")
test_dir = os.path.join(dataset_dir, "test")
```

### Load Data Example

```python
train_eggshell_dir = os.path.join(train_dir, "cangkang-telur")
print("Sample eggshell image:")
plt.imshow(load_img(f"{os.path.join(train_eggshell_dir, os.listdir(train_eggshell_dir)[np.random.randint(0, 500)])}"))
plt.show()

train_bottle_dir = os.path.join(train_dir, "botol plastik")
print("Sample bottle image:")
plt.imshow(load_img(f"{os.path.join(train_bottle_dir, os.listdir(train_bottle_dir)[np.random.randint(0, 500)])}"))
plt.show()
```

### Generate the Dataset for Usage

```python
train_generator, val_generator, test_generator = train_val_test_generator(train_dir, val_dir, test_dir)
```
