```python
#formating image
import PIL
import os.path
from PIL import Image, ImageFile

ImageFile.LOAD_TRUNCATED_IMAGES = True

f = r'/content/Dataset/dataset/'
for folder in os.listdir(f):
  os.mkdir('/content/formated_image/'+folder)
  print(folder)
  for file in os.listdir(f+folder):
    print(file)
    try:
      f_img = f+folder+'/'+file
      img = Image.open(f_img)
      img = img.convert("RGB")
      img = img.resize((300,300))
      if file.endswith(".jpeg") or file.endswith(".webp"): 
        file_name, file_ext = os.path.splitext(file)
      img.save('/content/formated_image/'+folder+'/{}.png'.format(file_name))
    except:
      pass
```
